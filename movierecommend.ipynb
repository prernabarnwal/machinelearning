```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('movies.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
dff=['genres', 'keywords', 'tagline', 'cast', 'director']
dff
```

```python
#df.isnull().sum()
```

```python
for feature in dff:
    df[feature]=df[feature].fillna('')
```

```python
combine=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
print(combine)
```

```python
vectorizer= TfidfVectorizer()
```

```python
feature_vectors= vectorizer.fit_transform(combine)
#print(feature_vectors)
```

# cosine similarity

```python
#getting the similarity scores using cosine similarity
similarity=cosine_similarity(feature_vectors)
#print(similarity)
#similarity.shape
```

```python
#getting movie name from user
movie_name=input('Enter your favourite movie name: ') #kiss of death, alive
```

```python
title_list= df['title'].tolist()
#print(title_list)
```

```python
#find close match for the movie name given by user
close_match=difflib.get_close_matches(movie_name, title_list)
#print(close_match)
```

```python
close_matched=close_match[0]
#close_matched
```

```python
#finding index of the movie with that title
index_movie=df[df.title==close_matched]['index'].values[0]
#index_movie
```

```python
#getting a list if similar movies
similarity_score= list(enumerate(similarity[index_movie]))
#print(similarity_score)
```

```python
#sorting the movies based on the similarity score
sorted_similar=sorted(similarity_score, key= lambda x:x[1], reverse=True)
#sorted_similar
```

```python
#print the name of similar movies based on index
print("Movies Recommended Based On Ur Favourite Movie")
i=1
for movie in sorted_similar:
    index=movie[0]
    title= df[df.index==index]['title'].values[0]
    if(i<21):
        print(i,'.',title)
        i+=1
```
